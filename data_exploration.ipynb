```python
import re
for c in temperatures_c:
    if re.compile('South').match(c): print(c)
    if re.compile('Slovak').match(c): print(c)
    if re.compile('North').match(c): print(c)        
```

'Slovak Republic' matches 'Slovakia', and we'll asume 'Korea' matches 'South Korea', as it is very unlike that the providers from the industry data had information about North Korean industries.

## 5. Check emissions and temperatures join

```python
print(len(temperatures.Country.unique()))
temperatures.Country.unique()
```

```python
print(len(emissions.Entity.unique()))
emissions.Entity.unique()
```

At first glance, it should work well. Let's try to make the join like in sql.py.

```python
import math
countries = emissions[["Entity", "Code"]].copy()
countries = countries.drop_duplicates()
countries = countries.dropna()
countries.columns = ["Country", "Code"]
countries = countries.set_index("Country")
countries
```

```python
join = temperatures.join(countries, on="Country", how="left")
join
```

```python
join[["Country", "Code"]].drop_duplicates().dropna().T
```

We have 188 countries, it worked really well! We should be able to join the tables on redshift without problems. :-)
